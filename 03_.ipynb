```python
from matplotlib import pyplot as plt    
import random
import numpy as np
```

```python
coin_tosses = np.random.binomial(n=10, p=0.5, size=1000)

values, counts = np.unique(coin_tosses, return_counts=True)

#print((values, counts))

probabilities = counts/1000

print(values, probabilities)
```

```python
plt.hist(coin_tosses)
plt.show()
```

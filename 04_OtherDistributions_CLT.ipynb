```python
import numpy as np
from matplotlib import pyplot as plt
```

## Normal Distribution

```python
mean = 70
sample_count = 1000000
standard_deviation = 7 # Roll number is used as the standard deviation
normal_samples = np.random.normal(mean, standard_deviation, size=sample_count)
plt.figure()
plt.hist(normal_samples, bins='auto', color ='blue', edgecolor = 'white', label='Normal Distribution')
plt.title("Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.grid(linewidth=0.4, alpha=0.7)
plt.legend()
plt.show()
```

## CLT

### For Normal Distribution

```python
from scipy.stats import norm

mean = 70
sample_count = 100
standard_deviation = 7 # Roll number is used as the standard deviation
normal_samples = np.random.normal(mean, standard_deviation, size=sample_count)

sample_size = 5
sample_means = []

for i in range (len(normal_samples)):
    sample = np.random.choice(normal_samples, sample_size, replace=True)
    sample_means.append(sample.mean())

plt.figure()

plt.hist(sample_means, bins='auto', density=True)

mean = np.mean(sample_means)
sigma = np.std(sample_means)

X = np.linspace(min(sample_means), max(sample_means), 1000)
Y = norm.pdf(X, mean, sigma)

plt.plot(X, Y)
```

But if we increase the number of samples, CLT jumps in and we get a more close to Normal distribution.

```python
from scipy.stats import norm

mean = 70
sample_count = 10000
standard_deviation = 7 # Roll number is used as the standard deviation
normal_samples = np.random.normal(mean, standard_deviation, size=sample_count)

sample_size = 500
sample_means = []

for i in range (len(normal_samples)):
    sample = np.random.choice(normal_samples, sample_size, replace=True)
    sample_means.append(sample.mean())

plt.figure()

plt.hist(sample_means, bins='auto', density=True)

mean = np.mean(sample_means)
sigma = np.std(sample_means)

X = np.linspace(min(sample_means), max(sample_means), 1000)
Y = norm.pdf(X, mean, sigma)

plt.plot(X, Y)
```

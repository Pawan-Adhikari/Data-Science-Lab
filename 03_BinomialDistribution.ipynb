```python
from matplotlib import pyplot as plt    
import numpy as np
```

### Bernoulli Distribution

```python
trials = int(1e7)
prob_success = 0.5

coin_tosses = np.random.binomial(1, p=prob_success, size=trials) #Simulating a single coin toss experiment.

values, counts = np.unique(coin_tosses, return_counts=True)

probabilities = counts/trials

print(values.astype(int), probabilities)
```

```python
plt.bar(values, probabilities, 0.1)
plt.xticks(np.arange(min(values), max(values)+1, 1)) #Discrete X axis, so integers work better.
plt.show()
```

## Binomial Distribution

```python
trials = int(1e7)
bernoulli_trials = 20
prob_success = 0.5

coin_tosses = np.random.binomial(n=bernoulli_trials, p=prob_success, size=trials)

values, counts = np.unique(coin_tosses, return_counts=True)

#print((values, counts))

probabilities = counts/trials

print(values.astype(int), probabilities)
```

```python
plt.bar(values, probabilities, 0.1)
plt.xticks(np.arange(min(values), max(values)+1, 1)) #Discrete X axis, so integers work better.
plt.show()
```

We are working with discrete random variables and thus discrete distribution. So, our distribution is best represented by a bar graph with ~ 0 width. 

```python
plt.plot(values, probabilities, 0.1)
plt.xticks(np.arange(min(values), max(values)+1, 1)) #Discrete X axis, so integers work better.
plt.show()
```

If the points are connected, the binomial distribution starts to take shape of a normal distribution. This is because with larger number of independent trials, a binomial distribution approaches a normal distribution, given by Central Limit Theorem. 

```python
trials = int(1e7)
bernoulli_trials = 1000 #Simulating a Large Large number of trials.
prob_success = 0.5

coin_tosses = np.random.binomial(n=bernoulli_trials, p=prob_success, size=trials)

values, counts = np.unique(coin_tosses, return_counts=True)

#print((values, counts))

probabilities = counts/trials

#print(values.astype(int), probabilities)

plt.bar(values, probabilities, 0.1) 
plt.show()
```

```python
plt.plot(values, probabilities, 0.1) 
plt.show()
```

## Biased Coin Bernoulli Distribution

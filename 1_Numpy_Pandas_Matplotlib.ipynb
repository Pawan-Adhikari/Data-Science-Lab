2D array, every column mean, subtract each element of the column with its mean.

```python
TwoD = np.array([[2,4,5],[6,7,8],[9,10,11]])
TwoD
```

```python
TwoD[:, 0]
```

```python
TwoD.shape[0]
```

```python
nNormalised_TwoD = TwoD - TwoD.mean(axis = 1)
nNormalised_TwoD
```

```python
Normalised_TwoD = []
for column in range(TwoD.shape[1]):
    current_column = TwoD[:, column]
    current_column_mean = current_column.mean()
    for index, values in enumerate(current_column):
        current_column[index] -= current_column_mean
    Normalised_TwoD = np.column_stack((Normalised_TwoD, current_column))
    #np.concatenate((Normalised_TwoD, current_column), axis=1)

Normalised_TwoD
```

3D plot:

```python
X = np.arange(0,10,0.5)
Y = np.arange(10,20,0.5)
X,Y = np.meshgrid(X,Y)
Z = np.cos(np.sqrt(X**2 + Y**2))

fig, ax = plt.subplots(subplot_kw={"projection":"3d"})
ax.plot_surface(X, Y, Z)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

X = np.arange(0,10,0.5)
Y = np.arange(10,20,0.5)
X,Y = np.meshgrid(X,Y)
Z = np.cos(np.sqrt(X**2 + Y**2))
levels = np.linspace(Z.min(), Z.max(), 7)

fig, ax = plt.subplots()
ax.contourf(X, Y, Z, levels = 25, cmap = 'viridis')
plt.show()
```
